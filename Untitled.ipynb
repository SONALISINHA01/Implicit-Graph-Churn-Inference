```python
# processing dataset
df = pd.read_csv('Telco-Customer-Churn.csv')

# Fix TotalCharges
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Drop customerID
df = df.drop(['customerID'], axis=1)


# ========== Churn ==========
df_churn = pd.get_dummies(df["Churn"], prefix="Churn", dtype=int)
df = pd.concat([df, df_churn], axis=1)
df = df.drop(['Churn', 'Churn_No'], axis=1)


# ========== Gender ==========
df_gender = pd.get_dummies(df["gender"], dtype=int)
df = pd.concat([df, df_gender], axis=1)
df = df.drop(['gender', 'Male'], axis=1)


# ========== Partner ==========
df_partner = pd.get_dummies(df["Partner"], prefix="Partner", dtype=int)
df = pd.concat([df, df_partner], axis=1)
df = df.drop(['Partner', 'Partner_No'], axis=1)


# ========== Dependents ==========
df_dep = pd.get_dummies(df["Dependents"], prefix="Dependents", dtype=int)
df = pd.concat([df, df_dep], axis=1)
df = df.drop(['Dependents', 'Dependents_No'], axis=1)


# ========== PhoneService ==========
df_phone = pd.get_dummies(df["PhoneService"], prefix="PhoneService", dtype=int)
df = pd.concat([df, df_phone], axis=1)
df = df.drop(['PhoneService', 'PhoneService_No'], axis=1)


# ========== MultipleLines ==========
df_mult = pd.get_dummies(df["MultipleLines"], prefix="MultipleLines", dtype=int)
df = pd.concat([df, df_mult], axis=1)
df = df.drop(['MultipleLines', 'MultipleLines_No'], axis=1)


# ========== InternetService ==========
df_internet = pd.get_dummies(df["InternetService"], prefix="InternetService", dtype=int)
df = pd.concat([df, df_internet], axis=1)
df = df.drop(['InternetService', 'InternetService_DSL'], axis=1)


# ========== OnlineSecurity / Backup / DeviceProtection / TechSupport / Streaming ==========
cols = ["OnlineSecurity", "OnlineBackup", "DeviceProtection",
        "TechSupport", "StreamingTV", "StreamingMovies"]

for c in cols:
    dummy = pd.get_dummies(df[c], prefix=c, dtype=int)
    df = pd.concat([df, dummy], axis=1)
    df = df.drop([c, f"{c}_No"], axis=1)


# ========== Contract ==========
df_contract = pd.get_dummies(df["Contract"], prefix="Contract", dtype=int)
df = pd.concat([df, df_contract], axis=1)
df = df.drop(['Contract', 'Contract_Month-to-month'], axis=1)


# ========== PaperlessBilling ==========
df_paper = pd.get_dummies(df["PaperlessBilling"], prefix="PaperlessBilling", dtype=int)
df = pd.concat([df, df_paper], axis=1)
df = df.drop(['PaperlessBilling', 'PaperlessBilling_No'], axis=1)


# ========== PaymentMethod ==========
df_payment = pd.get_dummies(df["PaymentMethod"], prefix="PaymentMethod", dtype=int)
df = pd.concat([df, df_payment], axis=1)
df = df.drop(['PaymentMethod','PaymentMethod_Bank transfer (automatic)'], axis=1)
```

```python
print(df.columns.tolist())
```

```python
y = df['Churn_Yes']

X = df.drop([
    'Churn_Yes',
    'MultipleLines_No phone service',
    'InternetService_No',
    'OnlineSecurity_No internet service',
    'OnlineBackup_No internet service',
    'DeviceProtection_No internet service',
    'TechSupport_No internet service',
    'StreamingTV_No internet service',
    'StreamingMovies_No internet service',
    'PaymentMethod_Credit card (automatic)'
], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# Scaling (important for many models)
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
log_reg = LogisticRegression(max_iter=1000)

# Train
log_reg.fit(X_train_scaled, y_train)

# Predict
y_pred = log_reg.predict(X_test_scaled)
y_prob = log_reg.predict_proba(X_test_scaled)[:,1]

# Metrics
log_acc = accuracy_score(y_test, y_pred)
log_auc = roc_auc_score(y_test, y_prob)

print("Logistic Regression Accuracy:", log_acc)
print("Logistic Regression ROC-AUC:", log_auc)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
dt = DecisionTreeClassifier(random_state=42)

# Train
dt.fit(X_train_scaled, y_train)

# Predict
y_pred = dt.predict(X_test_scaled)
y_prob = dt.predict_proba(X_test_scaled)[:,1]

# Metrics
dt_acc = accuracy_score(y_test, y_pred)
dt_auc = roc_auc_score(y_test, y_prob)

print("Decision Tree Accuracy:", dt_acc)
print("Decision Tree ROC-AUC:", dt_auc)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Create model
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    max_depth=None
)

# Train
rf.fit(X_train_scaled, y_train)

# Predict
y_pred = rf.predict(X_test_scaled)
y_prob = rf.predict_proba(X_test_scaled)[:,1]

# Metrics
rf_acc = accuracy_score(y_test, y_pred)
rf_auc = roc_auc_score(y_test, y_prob)

print("Random Forest Accuracy:", rf_acc)
print("Random Forest ROC-AUC:", rf_auc)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python

# Create model
gb = GradientBoostingClassifier(random_state=42)

# Train
gb.fit(X_train_scaled, y_train)

# Predict
y_pred = gb.predict(X_test_scaled)
y_prob = gb.predict_proba(X_test_scaled)[:,1]

# Metrics
gb_acc = accuracy_score(y_test, y_pred)
gb_auc = roc_auc_score(y_test, y_prob)

print("Gradient Boosting Accuracy:", gb_acc)
print("Gradient Boosting ROC-AUC:", gb_auc)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
xgb = XGBClassifier(
    eval_metric='logloss',
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    random_state=42
)

# Train
xgb.fit(X_train_scaled, y_train)

# Predict
y_pred = xgb.predict(X_test_scaled)
y_prob = xgb.predict_proba(X_test_scaled)[:,1]

# Metrics
xgb_acc = accuracy_score(y_test, y_pred)
xgb_auc = roc_auc_score(y_test, y_prob)

print("XGBoost Accuracy:", xgb_acc)
print("XGBoost ROC-AUC:", xgb_auc)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Create model
svm_model = SVC(probability=True, kernel='rbf', random_state=42)

# Train
svm_model.fit(X_train_scaled, y_train)

# Predict
y_pred = svm_model.predict(X_test_scaled)
y_prob = svm_model.predict_proba(X_test_scaled)[:,1]

# Metrics
svm_acc = accuracy_score(y_test, y_pred)
svm_auc = roc_auc_score(y_test, y_prob)

print("SVM Accuracy:", svm_acc)
print("SVM ROC-AUC:", svm_auc)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
